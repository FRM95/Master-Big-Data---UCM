```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Cargamos el dataset
df = pd.read_csv(
    r'datasets/Info_pais.csv',
    encoding='ISO-8859-1',
    delimiter=';'
)
```

```python
#Ordenacion del dataset
df_order = df.sort_values(
    'Esperanza de vida',
    ascending = True
)
```

```python
#Obtenemos los 5 primeros elementos ordenados por esperanza de vida
df_order.head()
```

```python
#Representacion de la renta per capita por esperanza de vida
plt.scatter(
    df_order['Renta per capita'], #eje x
    df_order['Esperanza de vida'] # eje y
);
```

```python
#Creacion de titulo y etiquetas
plt.title("Renta per capita ~ Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
plt.scatter(
    df_order['Renta per capita'], #eje x
    df_order['Esperanza de vida'] # eje y
);
```

```python
#Creacion de una nueva columna 'Poblacion normalizada' en el dataset
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / (max(df_order['Poblacion'])/10000)
```

```python
df_order.head()
```

```python
plt.title("Renta per capita ~ Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
plt.scatter(
    df_order['Renta per capita'], 
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'] #size
);
```

```python
#Modificamos el tamaño del grafico a la figura creada
plt.title("Renta per capita ~ Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
plt.scatter(
    df_order['Renta per capita'], 
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'] #size
);

fig = plt.gcf()
fig.set_size_inches(14.5, 10)
```

```python
#Añadimos colores en la representacion
plt.title("Renta per capita ~ Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
plt.scatter(
    df_order['Renta per capita'], 
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada'] #establecemos un color para el valor de la poblacion normalizada
);

fig = plt.gcf()
fig.set_size_inches(14.5, 10)
```

```python
plt.title("Renta per capita ~ Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
plt.scatter(
    df_order['Renta per capita'], 
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada'] 
);

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

#Bucle for para añadir las etiquetas
for i in range(0,5):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i], df_order['Esperanza de vida'][i])
    )
```

```python
plt.title("Renta per capita ~ Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
plt.scatter(
    df_order['Renta per capita'], 
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada'] 
);

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

for i in range(0,5):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i], df_order['Esperanza de vida'][i])
    )

#Mostramos los datos en el eje "y" desde el 1 al 120 de 10 en 10 escalado
plt.yticks(ticks=range(1,120,10));
```

```python
#Añadimos transparencia para poder observar los datos 

plt.title("Renta per capita ~ Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
plt.scatter(
    df_order['Renta per capita'], 
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada'],
    alpha = 0.5
);

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

for i in range(0,5):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i], df_order['Esperanza de vida'][i])
    )

plt.yticks(ticks=range(1,120,10));
```
